```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
from random import randint
import math
```

```python
# 5.
# Make a new Jupyter notebook named big_o_notation.ipynb

# Title your chart "Big O Notation"
# Label your x axis "Elements"
# Label your y axis "Operations"
# Label your curves or make a legend for the curves
# Use LaTex notation where possible

#note
# Certain characters require special escaping in TeX, such as:
# # $ % & ~ _ ^ \ { } \( \) \[ \]

fig = plt.figure(figsize=(12, 12))
axes = fig.subplots(nrows=4, ncols=2)
plt.style.use('seaborn')
fig.suptitle(r'$\alpha_i < \beta_i$')

n = range(-10, 11)


# Curves to graph

# y=0n+1 and label the curve "O(1)"

y = [(0*num+1) for num in n]
axes[0, 0].plot(n, y, color = 'red', label = '0(1)' )

# y=log(n) and label the curve "O(log n)"

y = [np.log(num) for num in n]
axes[0, 1].scatter(n, y, color = 'blue', label = '0(log n)', s=20)


# y=n and label the curve "O(n)"

y = n
axes[1, 0].plot(n, y, color = 'black', lw = '1.5', marker='o', markersize=5, label = '0(n)')


# y=nlog(n) and label it "O(n log n)"

y = [(num * np.log(num)) for num in n]
axes[1, 1].scatter(n, y, s=10, color='orange', label='O(n log n)')

# y=n2 and label it "O(n^2)"

y = [(num**2) for num in n]
axes[2, 0].scatter(n, y, s=5, color = 'magenta', label = 'O(n^2)')

# y=2n and label it "O(2^n)"

y = [(2**num) for num in n]
axes[2, 1].plot(n, y, color = 'brown', lw='1', label = 'O(2^n)')

# y=n! and label it "O(n!)"

y = [(np.math.factorial(np.abs(num))) for num in n]
axes[3, 0].scatter(n, y, s=4, color = 'gray', label = 'O(n!)')


# y=nn and label it "O(n^n)"

y = [(num**num) for num in n]
axes[3, 1].plot(n, y, color = 'violet', lw='1', label = 'O(n^n)')



lines = []
labels = []

for ax in fig.axes:
    axLine, axLabel = ax.get_legend_handles_labels()
    lines.extend(axLine)
    labels.extend(axLabel)


fig.legend(lines, labels, loc='upper center', ncol=len(labels))
fig.tight_layout()
```
